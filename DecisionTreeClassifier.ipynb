# Decision Tree Classifier in Python using Scikit-learn
### Decision Trees can be used as classifier or regression models.


### Entropy
The root node (the first decision node) partitions the data using the feature that provides the most information gain.

Information gain tells us how important a given attribute of the feature vectors is.

It is calculated as:

### Information Gain=entropy(parent)–[average entropy(children)]
Where entropy is a common measure of target class impurity, given as:

### Entropy=Σi–pi log2 pi
where i is each of the target classes.

### Gini Impurity
Gini Impurity is another measure of impurity and is calculated as follows:

### Gini=1–Σi p2i
Gini impurity is computationally faster as it doesn’t require calculating logarithmic functions, though in reality which of the two methods is used rarely makes too much of a difference.

```python
# Load libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from IPython.display import Image  
from sklearn import tree
import pydotplus
#https://towardsdatascience.com/random-forest-in-python-24d0893d51c0
#datasets : https://vincentarelbundock.github.io/Rdatasets/datasets.html
```

```python
balance_data = pd.read_csv(
'https://archive.ics.uci.edu/ml/machine-learning-databases/balance-scale/balance-scale.data',
                           sep= ',', header= None)
```

```python
balance_data[:5]
```

```python
X = balance_data.values[:, 1:5]
Y = balance_data.values[:,0]
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 100)
```

## Decision Tree Classifier with criterion gini index

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_gini.fit(X_train, y_train)
```

```python
balance_data.columns
column = []
for digit in balance_data.columns[1:5]:
    column.append (str(digit))
column1=np.array(column)
```

```python
uni=balance_data[0].unique()
uni
```

```python
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
dot_data = tree.export_graphviz(clf_gini, out_file=None, 
                                feature_names=column1,  
                                class_names=uni)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
```

## Decision Tree Classifier with criterion as information gain

```python
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
 max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
dot_data = tree.export_graphviz(clf_entropy, out_file=None, 
                                feature_names=column1,  
                                class_names=uni)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
```

## Prediction

```python
y_pred_gini = clf_gini.predict(X_test)
y_pred_gini
```

```python
y_pred_en = clf_entropy.predict(X_test)
y_pred_en
```

## Calculating Accuracy Score

#### Accuracy for Decision Tree classifier with criterion as gini index

```python
from sklearn.metrics import accuracy_score
print("Accuracy is ", accuracy_score(y_test,y_pred_gini)*100)
```

#### Accuracy for Decision Tree classifier with criterion as information gain

```python
print("Accuracy is ", accuracy_score(y_test,y_pred_en)*100)
```
